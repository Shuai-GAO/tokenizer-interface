```python
def getChinese(context):
#     context = context.decode("utf-8") # convert context from str to unicode
    filtrate = re.compile(u'[^\u4E00-\u9FA5]') # non-Chinese unicode range
    context = filtrate.sub(r'', context) # remove all non-Chinese characters
#     context = context.encode("utf-8") # convert unicode back to str
    return context

from bisect import bisect_left

def binary_search(sorted_list,token, lo=0, hi=None):   # can't use a to specify default for hi
    hi = hi if hi is not None else len(sorted_list) # hi defaults to len(a)
    pos = bisect_left(sorted_list,token,lo,hi)          # find insertion position
    return (pos if pos != hi and sorted_list[pos] == token else -1) # don't walk off the end
```
